```python
#Import Libraries
import numpy as np
import pandas as pd
import yfinance as yf
from datetime import datetime, timedelta
import pickle
import time
from typing import Optional, Dict, List
import functools
import requests
import warnings
from pathlib import Path
warnings.filterwarnings('ignore')
#Quick Test
print(f'Timestamp: {datetime.now()}')
```

```python
#Cache Directory
def get_cache(key, fetch, ttl_hours=1):
    cache_file=Path(f'./cache/{key}.pkl')

    #Check cache 
    if cache_file.exists():
        age=datetime.now() - datetime.fromtimestamp(cache_file.stat().st_mtime)
        if age<timedelta(hours=ttl_hours):
            print(f'Loading {key} from cache')
            return pickle.load(open(cache_file, 'rb'))

    #Fetch data if bad cache
    print(f'Fetching new {key}...')
    data=fetch()

    #Save
    Path('./cache').mkdir(exist_ok=True)
    pickle.dump(data, open(cache_file, 'wb'))

    return data

    
```

```python
# Market Data Class
class MarketDataFetcher:
    def __init__(self,ttl_hours=1):
        self.ttl_hours=ttl_hours
    @staticmethod
    def atlas_cache(func):
        @functools.wraps(func)
        def wrapper(self, *args, **kwargs):
            all_args=list(args) + [f'{k}={v}' for k, v in kwargs.items()]
            cache_key=f'{func.__name__}_{'_'.join(map(str, all_args))}'

            return get_cache(cache_key, lambda: func(self,*args,**kwargs), self.ttl_hours)
        return wrapper
    
    @atlas_cache
    def get_equity_price(self, ticker, period='1mo'):
        """
        Fetch historical equity prices using yfinance
        Use get_cache to avoid repeated API calls
        
        Args:
            ticker: Stock symbol (e.g., "SPY", "AAPL")
            period: Time period ("1d", "5d", "1mo", "3mo", "1y")
        
        Returns:
            DataFrame with OHLCV data
        """
        return yf.download(ticker, period=period, progress=False)

    @atlas_cache
    def get_current_spot(self, ticker):
        """
        Fetch the current price using yfincance

        Returns:
            Float: current price
        """
        try:
            # yf.download is generally more stable than yf.Ticker
            data = yf.download(ticker, period="1d", progress=False)
            if data.empty:
                # Fallback to a hardcoded price so the rest of your code doesn't crash
                print(f"âš ï¸ Yahoo blocked {ticker}. Using fallback price.")
                return 500.0 if ticker == "SPY" else 150.0
            
            # Using .iloc[-1] to get the most recent closing price
            return float(data['Close'].iloc[-1])
        except Exception as e:
            print(f"âš ï¸ Error fetching {ticker}: {e}. Using fallback.")
            return 500.0 # Default SPY price
    @atlas_cache
    def get_option_chain(self, ticker, n_expirations=5):
        """
        Fetch option chains for the first n expirations
        """
        stock = yf.Ticker(ticker)
        all_exps= stock.options
        exps_to_fetch = all_exps[0:n_expirations]

        chains = {}
        for exp in exps_to_fetch:
            opt = stock.option_chain(exp)
            chains[exp]={'calls':opt.calls,'puts':opt.puts}
        return chains

    @atlas_cache
    def get_treasury_curve(self):
        """
        Fetch Treasury yields for the risk free rate curve
        """
        symbols={'0.25':'^IRX', '5.0':'^FVX','10':'^TNX'}
        curve={}

        for maturity, ticker in symbols.items():
            yield_val=self.get_current_spot(ticker)/100

            curve[float(maturity)]=yield_val
        return pd.Series(curve).sort_index()
        
            
        
```

```python
#Option Processor
class OptionProcessor:
    def __init__(self, spot_price):
        self.spot_price = spot_price

    def clean_chain(self, chain_df):
        """
        Cleans a single data frame of puts or calls
        """
        df= chain_df.copy()

        #Calculate Mid Price
        df['mid_price']= (df['bid']+df['ask']) /2
        #Filter Out Bad Data
        df= df[df['mid_price']>0].dropna(subset=['mid_price'])
        df= df[df['volume']>0]
        df['moneyness']=df['strike'] / self.spot_price
        df=df[(df['moneyness']>0.8) & (df['moneyness']<1.2)]

        return df

    def prepare_for_calibration(self, raw_chains):
        """
        Takes raw_chains dict from MarketDataFetcher and returns one clean DataFrame
        """
        all_data=[]

        for expiry, data in raw_chains.items():
            #Process and cleans calls & puts
            calls=self.clean_chain(data['calls'])
            puts=self.clean_chain(data['puts'])

            #Calculate time to maturity
            today=datetime.now()
            expiry_dt = datetime.strptime(expiry, '%Y-%m-%d')
            time_to_mat= (expiry_dt-today).days/365.25

            #Add to metadata
            for df, opt_type in zip([calls,puts],['call','put']):
                df['T']=time_to_mat
                df['type']=opt_type
                all_data.append(df)

        #Combine
        return pd.concat(all_data).reset_index(drop=True)  

            

        
```

```python
# ==========================================
# ATLAS PIPELINE TEST: SPY VOLATILITY SMILE
# ==========================================

try:
    # 1. Initialize
    fetcher = MarketDataFetcher()
    
    # 2. Ingest
    print("ğŸš€ Step 1: Ingesting data...")
    ticker = "SPY"
    spot = fetcher.get_current_spot(ticker)
    raw_chains = fetcher.get_option_chain(ticker, n_expirations=3)
    curve = fetcher.get_treasury_curve()
    
    # 3. Process
    print("ğŸ› ï¸ Step 2: Processing options...")
    processor = OptionProcessor(spot_price=spot)
    clean_df = processor.prepare_for_calibration(raw_chains)
    
    # 4. Results
    print("\n" + "="*30)
    print("      ATLAS TEST REPORT")
    print("="*30)
    print(f"Ticker:           {ticker}")
    print(f"Spot Price:       ${spot:.2f}")
    print(f"Risk-Free Rate:   {curve['10.0']:.2%}") # 10Y Yield
    print(f"Raw Expirations:  {len(raw_chains)}")
    print(f"Clean Data Rows:  {len(clean_df)}")
    
    # Check the "Smile" logic
    min_m = clean_df['moneyness'].min()
    max_m = clean_df['moneyness'].max()
    print(f"Moneyness Range:  {min_m:.2f} to {max_m:.2f}")
    
    print("\n--- Sample of Processed Data ---")
    print(clean_df[['strike', 'mid_price', 'T', 'type', 'moneyness']].head())
    
    # 5. Sanity Check
    assert 0.79 < min_m < 1.21, "Moneyness filter failed!"
    assert clean_df['mid_price'].min() > 0, "Zero price detected in clean data!"
    print("\nâœ… PIPELINE TEST PASSED: Foundation is solid.")

except Exception as e:
    print(f"\nâŒ PIPELINE TEST FAILED")
    print(f"Error Type: {type(e).__name__}")
    print(f"Message:    {e}")
```
